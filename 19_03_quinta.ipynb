Desenvolva um programa que solicite do usuário um valor, a partir desse, calcule o Fatorial e mostre. Solicite o valor até o usuário informar um número negativo.

```python
def fatorial(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

valor = 0

while valor >= 0:
    valor = eval(input("Digite um valor: "))

    if (valor < 0):
        print("Você digitou um número negativo")
        break

    fatorial_valor = fatorial(valor)
    print("Fatorial do valor digitado:", fatorial_valor)
```

Solicite Notas de alunos para seu usuário até informar um número negativo. No término mostrar quantos alunos tiraram nota 5 e quantos tiraram nota 8.

```python
a = 0
qtd_alunos = 0
notas = []

while a > -1:
    nota = int(input(f"Digite a nota do aluno {a}: "))

    if (nota < 0):
        break

    notas.append(nota)
    a += 1
    qtd_alunos += 1

print(f"Alunos com notas: {qtd_alunos}")
print(f"Alunos nota 5: {notas.count(5)}")
print(f"Alunos nota 8: {notas.count(8)}")
```
